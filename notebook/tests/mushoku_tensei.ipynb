```python
import requests
import os
from bs4 import BeautifulSoup

ROOT = os.path.dirname(os.path.abspath("")) + "/recon/"
print(ROOT)
pages = 90
downloads_folder = "downloads/"
if not os.path.isdir(ROOT + downloads_folder):
    os.mkdir(ROOT + downloads_folder)

pages = []
main = requests.get("https://mushokutenseiscan.com/")
soup = BeautifulSoup(main.content, "html.parser")
for link in soup.find_all('a'):
    if "-chapter-" in link["href"]:
        pages.append(link["href"])

for page in pages:
    print(f"Processing page {page}...")
    folder_name = page.replace("https://mushokutenseiscan.com/mushoku-tensei-chapter-", "")
    folder = ROOT + downloads_folder + folder_name
    if not os.path.isdir(folder):
        print(folder)
        os.mkdir(folder)
        r = requests.get(page)
        html = r.content
        soup = BeautifulSoup(html, "html.parser")
        for link in soup.find_all('img',{'class':'aligncenter'}):
            image_url = link["src"]
            image_name = os.path.basename(image_url)
            image_data = requests.get(image_url)
            with open(folder + image_name, "wb") as img:
                img.write(image_data.content)
```

```python
import zipfile
import re

def tryint(s):
    try:
        return int(s)
    except:
        return s

def alphanum_key(s):
    return [ tryint(c) for c in re.split('([0-9]+)', s) ]
    
with zipfile.ZipFile(f"MUSHOKU_TENSEI.cbz", mode="w") as archive:
    for chapter in os.listdir(downloads_folder):
        full_chapter = downloads_folder + chapter
        print(f"Processing chapter {chapter}...")
        images = os.listdir(full_chapter)
        images.sort(key=alphanum_key)
        for image in images:
            if image.endswith(".jpg"):
                archive.write(full_chapter + "/" + image, f"MUSHOKU TENSEI/Chapter {chapter}/" + os.path.basename(full_chapter + "/" + image))
                
```
